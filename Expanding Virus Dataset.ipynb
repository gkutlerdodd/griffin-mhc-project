```python
virus_dict = {}

with open("virushostdb.tsv") as file :
    next(file)
    for line in file :
        line = line.rstrip().split("\t")
        if line[0] not in virus_dict :
            virus_dict[line[0]] = {
                "name": line[1],
                "refseq_id": [x.lstrip() for x in line[3].split(",")]
            }
            
names = []
for key,value in virus_dict.items() :
    formatted_name = value["name"].replace(" ", "_").replace("/", "")
    names.append(formatted_name)

print(virus_dict)

#with open("viruses.csv", "w", newline="") as file :
    #wr = csv.writer(file, quoting=csv.QUOTE_ALL)
    #wr.writerow(names)
```

```python
import re

with open("virushostdb.cds.faa") as file :
    file_contents = file.read()
virus_seqs = re.findall("([>][^|]*[|][^|]*[|][^|]*[|]([^|]*)[|].*[^>]*)", file_contents)
```

```python
for key,value in virus_dict.items() :
    file = ""
    formatted_name = value["name"].replace(" ", "_").replace("/", "")
    file_name = formatted_name + ".fasta"
    for accession_no in value["refseq_id"] :
        for seq in virus_seqs :
            if seq[1] == accession_no :
                file += seq[0]
                
    with open(file_name, "w") as f :
        f.write(file)
```
